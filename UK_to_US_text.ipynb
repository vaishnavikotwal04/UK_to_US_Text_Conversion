```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Embedding, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

```

```python
# Load the dataset
file_path = '/content/CozmoX Assignment Dataset.csv'  # Change this to your file path
data = pd.read_csv(file_path)
```

```python
df
```

```python

# Extract input and target texts
input_texts = data['input_text'].tolist()  # UK English text (input)
target_texts = data['target_text'].tolist()  # US English text (target)

# Add <start> and <end> tokens to target texts for the decoder
target_texts_with_start = ['<start> ' + text for text in target_texts]
target_texts_with_end = [text + ' <end>' for text in target_texts_with_start]
```

```python

# Tokenize the texts
tokenizer_input = Tokenizer()
tokenizer_target = Tokenizer()

tokenizer_input.fit_on_texts(input_texts)
tokenizer_target.fit_on_texts(target_texts_with_end)

# Convert text to sequences (token IDs)
input_sequences = tokenizer_input.texts_to_sequences(input_texts)
target_sequences = tokenizer_target.texts_to_sequences(target_texts_with_end)

# Padding the sequences to the same length
max_len_input = max(len(seq) for seq in input_sequences)
max_len_target = max(len(seq) for seq in target_sequences)

input_sequences = pad_sequences(input_sequences, maxlen=max_len_input, padding='post')
target_sequences = pad_sequences(target_sequences, maxlen=max_len_target, padding='post')
```

```python

# Split the dataset for training and testing
X_train, X_test, y_train, y_test = train_test_split(input_sequences, target_sequences, test_size=0.2, random_state=42)
```

```python

# Vocabulary sizes
input_vocab_size = len(tokenizer_input.word_index) + 1
target_vocab_size = len(tokenizer_target.word_index) + 1
```

```python


# Define the model architecture
def build_model(input_vocab_size, target_vocab_size, max_len_input, max_len_target):
    # Encoder
    encoder_input = Input(shape=(max_len_input,))
    encoder_embedding = Embedding(input_vocab_size, 128)(encoder_input)
    encoder_lstm = LSTM(256, return_state=True)
    encoder_output, state_h, state_c = encoder_lstm(encoder_embedding)

    # Decoder
    decoder_input = Input(shape=(max_len_target,))
    decoder_embedding = Embedding(target_vocab_size, 128)(decoder_input)
    decoder_lstm = LSTM(256, return_sequences=True)(decoder_embedding, initial_state=[state_h, state_c])
    decoder_output = Dense(target_vocab_size, activation='softmax')(decoder_lstm)

    # Model
    model = Model([encoder_input, decoder_input], decoder_output)
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    return model
```

```python

# Build the model
model = build_model(input_vocab_size, target_vocab_size, max_len_input, max_len_target)

# Model summary
model.summary()

# Prepare the decoder input and output for training
decoder_input_train = np.zeros_like(y_train)
decoder_input_train[:, 1:] = y_train[:, :-1]  # Decoder input is y_train shifted by one position

# Train the model
model.fit([X_train, decoder_input_train], np.expand_dims(y_train, -1), epochs=10, batch_size=64, validation_split=0.1)

# Save the model
model.save('input_to_target_translation_model.h5')

```

```python
import numpy as np
from tensorflow.keras.models import load_model
import pickle
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load the trained model
model = load_model('/content/input_to_target_translation_model.h5')

# Load the tokenizers
with open('tokenizer_input.pkl', 'rb') as f:
    tokenizer_input = pickle.load(f)

with open('tokenizer_target.pkl', 'rb') as f:
    tokenizer_target = pickle.load(f)

# Set the maximum sequence lengths used during training
max_len_input = 4  # This should match the length used during training for input
max_len_target = 6  # This should match the length used during training for the target output

# Ensure that the <start> token exists in the target tokenizer
if '<start>' not in tokenizer_target.word_index:
    tokenizer_target.word_index['<start>'] = len(tokenizer_target.word_index) + 1
    tokenizer_target.index_word[tokenizer_target.word_index['<start>']] = '<start>'

# Function to convert input text to target text (UK to US translation)
def translate_input_to_target(input_text):
    # Tokenize and pad the input
    input_seq = tokenizer_input.texts_to_sequences([input_text])
    input_seq_padded = pad_sequences(input_seq, maxlen=max_len_input, padding='post')

    # Check if the sequence contains any out-of-vocabulary (OOV) token
    if any([word not in tokenizer_input.word_index for word in input_text.split()]):
        raise ValueError("Input contains out-of-vocabulary words.")

    # Initialize the decoder input, adding the <start> token at the beginning
    decoder_input = np.zeros((1, max_len_target))  # Initialize with zeros
    decoder_input[0, 0] = tokenizer_target.word_index['<start>']  # Add the <start> token

    # Predict the sequence
    predicted_sequence = model.predict([input_seq_padded, decoder_input])

    # Convert the predicted sequence to words
    predicted_ids = np.argmax(predicted_sequence[0], axis=-1)

    # Make sure that indices are within the valid range
    predicted_ids = [id for id in predicted_ids if id != 0]  # Remove padding tokens (ID=0)
    predicted_words = [tokenizer_target.index_word.get(id, '') for id in predicted_ids]

    return ' '.join(predicted_words)

# Test the model with an input text example
input_example = 'I like the colour blue'
try:
    target_translation = translate_input_to_target(input_example)
    print(f"Input Text: {input_example}")
    print(f"Target Translation: {target_translation}")
except ValueError as e:
    print(str(e))
```

```python
from transformers import pipeline

# Load the translation model for UK English to US English
translator = pipeline("translation_en_to_en", model="facebook/m2m100_418M")

def convert_uk_to_us(text):
    # Translate the text
    translation = translator(text, src_lang="en", tgt_lang="en")
    return translation[0]['translation_text']

# Input UK English text
uk_text = input("Enter UK English text: ")

# Convert to US English
us_text = convert_uk_to_us(uk_text)

print(f"Converted US Text: {us_text}")
```
